```python
import pandas as pd
from src.article_processor.article_processor import ArticleProcessor
from sklearn.metrics import classification_report
```

```python
train_labels = pd.read_pickle('../data/subset_train.pkl')
```

```python
def get_fts(entity_a, entity_b, article_id):
    return ArticleProcessor(article_id, entity_a, entity_b).features
```

```python
train_features = train_labels.apply(lambda x: get_fts(x.entity_a, x.entity_b, x.article_id), axis=1)
```

```python
train_labels['train_features'] = train_features
```

```python
%pwd
```

```python
train_labels.to_pickle('data/subset_train.pkl')
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV, SGDClassifier
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix
```

```python
train_labels = train_labels.dropna()
```

```python
%pwd
```

```python
train_labels
```

```python
train_labels.to_pickle('data/train_features.pkl')
```

```python
svm_pipeline = Pipeline(steps=[('dict_vect', DictVectorizer(sparse=False)), 
                       ('impute', SimpleImputer(fill_value=-1)),
                       ('std', StandardScaler()), 
                       ('norm', Normalizer()),
                       ('svm', SGDClassifier(class_weight='balanced'))])
```

```python
pipeline = Pipeline(steps=[('dict_vect', DictVectorizer(sparse=False)), 
                           ('impute', SimpleImputer(fill_value=-1)),
                           ('std', StandardScaler()), 
                           ('norm', Normalizer()),
                           ('lr', LogisticRegressionCV(max_iter=500,
                                                       class_weight='balance',
                                                       scoring='f1_micro',
                                                       solver='newton-cg'
                                                      ))])
```

```python
X = train_labels['train_features']
y = train_labels['relation']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.333)
```

```python
cross_val_score(pipeline, X, y, cv=5, scoring='f1_micro')
```

```python
pipeline.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
map_relations =  {'P22': 'father', 'P25': 'mother', 'P26': 'spouse', 'P3373': 'sibling', 'P40': 'child'}
```

```python
plt.rcParams["figure.figsize"] = (10,7)
```

```python
for key in map_relations.keys():
    name = map_relations[key]
    f1_micro = pipeline.named_steps['lr'].scores_[key].mean(axis=0)
    plt.plot(f1_micro)
plt.legend([f'micro_f1_{map_relations[key]}' for key in list(map_relations.keys())])
plt.title('Validation Micro-F1 Score During Training')
plt.xlabel('Fold Number')
plt.ylabel('Average Validation Micro-F1 Score')
plt.figure(figsize=(200,10))
plt.show()
```

```python
y_pred = pipeline.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
