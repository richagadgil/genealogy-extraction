```python
def check_entities_mentioned(entity_a_id, entity_b_id, article_id):
    article_tagged = wiki_referencer.get_article_tags(article_id)
    #print(article_tagged)
    check_a = re.search(f'@{entity_a_id}@', article_tagged)
    check_b = re.search(f'@{entity_b_id}@', article_tagged)
    if check_a and check_b:
        return True
    return False
```

```python
train_labels['contains_entities'] = train_labels.apply(lambda x: check_entities_mentioned(x.entity_a, x.entity_b, x.article_id), axis=1)
```

```python
subset_train = train_labels[train_labels['contains_entities'] == True]
```
